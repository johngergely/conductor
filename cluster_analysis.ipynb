```python
## wrap clustering algorithm

def runCluster(raw_data, n_clusters):
    n_features = 1
    n_samples = len(raw_data)
    data = np.array(raw_data)
    data = np.reshape(data, (n_samples,n_features))
    print "data with shape",np.shape(data),np.size(data)

    estimator = KMeans(n_clusters=n_clusters)
    estimator.fit(data)
    return estimator.labels_
```

```python
## hacked time correlation estimators
## not debugged yet

def corr(F):
    N = len(F)
    F = F - F.mean()
    g_min = 1
    g_max = int(N/2)
    gs = [g for g in range(g_min, g_max)]
    C_g = np.zeros((len(gs)),float)
    for g in gs:
        g_i = g - g_min
        C_g[g_i] = corr_g(F, g, norm=1.)
    return gs, C_g

def corr_g(F_dev,g,norm=1.):
    M = len(F_dev) - g
    return norm/M * np.sum(F_dev[0:-(g+1)] * F_dev[g:-1])
```

```python
## pull bad data out before analysis

print len(tTrip),"original records"
tTrip = tTrip.dropna(how='any')
## cut out unreasonable times that are artifacts of
## midnight crossings (these can be dealt with) and glitches in data collection
print len(tTrip),"records after removing nan values"
dt_max = 150*60.
dt_min = 45*60.
df_mask = (tTrip['trip_time']>dt_min) & (tTrip['trip_time']<dt_max)
tTrip = tTrip[df_mask]
print len(tTrip),"records after culling bad data"
#xval = tTrip['time_of_day'].values
#yval = tTrip['trip_time'].values/60.
```

```python
tTrip['line'] = tTrip['line'].astype('string')
tTrip['line'].dtype
```

```python
# try out KMeans clustering

bokeh_colors = ['tomato', 'yellow', 'turquoise', 'seagreen', 'purple', 'lime', 'gray', 'deeppink', 'black', 'red']
bokeh_shapes = {'4':'cross', '5':'circle', '6':'triangle'}

figure()
hold()

# this version does clustering on the whole set to see the close
# relationship between subway line and cluster affiliation
n_clusters=3
labels = runCluster(tTrip['trip_time'].values, n_clusters)
tTrip['labels'] = labels
colors = [bokeh_colors[i] for i in labels]
line_shapes = [bokeh_shapes[l] for l in tTrip['line'].values]

for l in ['4','5','6']:
    select_df = tTrip[tTrip['line']==l]
    scatter(select_df['time_of_day'].astype(float).values, 
            select_df['trip_time'].values/60., 
            color=[bokeh_colors[i] for i in select_df['labels'].values], 
            marker=bokeh_shapes[l])

show()
```

```python
# this version does clustering for each line separately, 
# plots them all together... not sure if there's any insight here

figure()
hold()

c_index = 0
for l in ['4','5','6']:
    n_clusters = 3
    select_df = tTrip[tTrip['line']==l]
    labels = runCluster(select_df['trip_time'].values, n_clusters)
    lab_colors = [bokeh_colors[(i+c_index)%len(bokeh_colors)] for i in labels]
    scatter(select_df['time_of_day'].astype(float).values, 
            select_df['trip_time'].values/60., 
            color=lab_colors, 
            marker=bokeh_shapes[l])
    c_index += n_clusters

show()
```
